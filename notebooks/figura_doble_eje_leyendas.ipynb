```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x   = np.linspace(-np.pi,np.pi, 100)
y1  = np.sin(x)
y2  = 100 * np.cos(x)
```

```python
fig, ax1 = plt.subplots(figsize=(10,4))

g1, = ax1.plot(x,y1,'r.',label = "Uno")
ax2 = ax1.twinx()
g2, = ax2.plot(x,y2,'b+',label = "Cien")

ax1.set_ylabel("Uno")
ax2.set_ylabel("Cien")

ax1.legend(loc="upper left")
ax2.legend()
```

```python
lineas = [g1, g2]
```

```python
etiquetas = [linea.get_label() for linea in lineas]
```
