```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("../dat/datos_ruoa_esolmet.csv",index_col=0,
                   parse_dates=True)
datos.columns
```

```python
datos[["Ig_esolmet","Ig_ruoa"]].plot(subplots=True)
```

```python
datos["2018-01":"2018-02"][["Ig_esolmet","Ig_ruoa"]].plot(subplots=True)
```

```python
datos[datos.index.month==11].Ig_esolmet.plot() 
## Ventaja de que esto aplica para multiples a~nos
```

```python
datos.index.dayofweek.unique()
```

```python
fines = datos[(datos.index.dayofweek>4) & (datos.index.month==11)]
fig, ax = plt.subplots(figsize=(20,8))

ax.plot(fines.Ig_esolmet,"r+")
ax.plot(fines.Ig_ruoa,"k.")
```

```python
fin2 = datos.truncate(before="2018-11-09",after="2018-11-10")

fig, ax = plt.subplots(figsize=(20,8))

ax.plot(fin2.Ig_esolmet,"r+")
ax.plot(fin2.Ig_ruoa,"k.")
```
