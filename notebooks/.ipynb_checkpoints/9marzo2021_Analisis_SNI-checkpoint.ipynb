```python
import pandas as pd
import seaborn as sns
```

```python
sni = pd.read_csv("../dat/SNI_Vigentes_2020.csv")
sni.head()
```

```python
len(sni)
```

```python
sni.columns
```

```python
sni.Nobilis.value_counts()
```

```python
datos = sni[(sni.Nobilis=="DRA." ) | (sni.Nobilis=="DR.")]
len(datos)/len(sni)
```

```python
sns.catplot(x="Nobilis",kind="count",data=datos,
           order=["DRA.","DR."])
```

```python
sns.catplot(x="Nobilis",kind="count",data=datos,
           order=["DRA.","DR."],row="Categoría",sharex=False);
```

```python
sns.catplot(x="Nobilis",kind="count",data=datos,
           order=["DRA.","DR."],row="Área del Conocimiento",sharex=False);
```
